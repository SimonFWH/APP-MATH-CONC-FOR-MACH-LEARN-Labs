The process below shows how to implement a simple linear regression analysis for a vehicle pricing scenario. Re-run it and take note of the R-sqaured value. Then. change the std value in np.random.normal to 1, 5 and 10. Observe the change in R-sqaured. How do you explain the change in R-sqaured value?

:: After experimenting on the std value in np.random.normal to different values. We observe that the R-squared value decreases immensely as the std increases. Similarly,the plots on the graph scatters more as the std increases, while the plots are more compressed to a slope when the value is lower.

Last but not least the R-square value is closest to 1 when std is 1, so it is the most fit model out of them under this scenario.

## when std is set to 1 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

vehicle_age = np.random.uniform(0,10,1000)
vehicle_price = 70000 - (vehicle_age + np.random.normal(0, 1, 1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price = vehicle_price.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price)
```

```python
reg = LinearRegression().fit(vehicle_age, vehicle_price)
reg.score(vehicle_age, vehicle_price)
```

## When std is set to 5

```python
vehicle_age = np.random.uniform(0,10,1000)
vehicle_price_2 = 70000 - (vehicle_age + np.random.normal(0, 5, 1000)) * 3000
vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price_2 = vehicle_price_2.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price_2)
```

```python
reg = LinearRegression().fit(vehicle_age, vehicle_price_2)
reg.score(vehicle_age, vehicle_price_2)
```
